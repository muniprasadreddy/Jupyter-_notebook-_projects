Problem Statement:Home Credit Default Risk is about predicting whether someone will repay a loan or not. By analyzing their financial history and other personal details, we can determine if they are a high or low risk for defaulting. This helps lenders decide who to approve for a loan.

```python
SK_ID_CURR: Unique identifier for each loan application.
TARGET: Indicates if the applicant had payment difficulties (1 = yes, 0 = no).
NAME_CONTRACT_TYPE: Identifies the type of loan contract.
CODE_GENDER: Gender of the applicant.
FLAG_OWN_CAR: Indicates if the applicant owns a car.
FLAG_OWN_REALTY: Indicates if the applicant owns a house or flat.
CNT_CHILDREN: Number of children the applicant has.
AMT_INCOME_TOTAL: Total annual income of the applicant.
AMT_CREDIT: Credit amount of the loan.
AMT_ANNUITY: Annual annuity amount of the loan.
AMT_GOODS_PRICE: For consumer loans, it's the price of the goods for which the loan is given.
NAME_TYPE_SUITE: Who accompanied the client when applying for the loan.
NAME_INCOME_TYPE: Client's income type (e.g., working, pensioner, etc.).
NAME_EDUCATION_TYPE: Highest education level of the client.
NAME_FAMILY_STATUS: Family status of the client.
NAME_HOUSING_TYPE: What type of housing the client lives in.
DAYS_BIRTH: Client's age in days at the time of application.
DAYS_EMPLOYED: How many days before the application the person started current employment.
FLAG_MOBIL: Is there a mobile phone linked to the client.
OCCUPATION_TYPE: Client's occupation.
CNT_FAM_MEMBERS: How many family members the client has.
Bureau Data (bureau.csv):SK_ID_BUREAU: Unique identifier for each credit bureau record.
SK_ID_CURR: Unique identifier for each loan application.
CREDIT_ACTIVE: The status of the credit (e.g., active, closed).
CREDIT_CURRENCY: The currency of the loan.
DAYS_CREDIT: How many days before the application the bureau credit was applied for.
CREDIT_DAY_OVERDUE: Number of days the credit line is overdue.
AMT_CREDIT_MAX_OVERDUE: Maximal amount overdue on the credit.
CNT_CREDIT_PROLONG: Number of prolongations of the credit.
```
