# matplotlip  in Barplot

```python
import matplotlib.pyplot as plt
import numpy as np
x=[1,2,3,4,5,6,7,8,9]
y=[20,30,40,50,60,70,80,90,100]

plt.xlabel('population',fontsize=15)
plt.ylabel('age',fontsize=15)
plt.title(' age of city popualation',fontsize=20)
c=['blue','green','pink','red',"maroon",'black','yellow']
plt.bar(x,y, color=c,edgecolor='red',alpha=0.5,label="age")
plt.legend()
plt.show()
```

# Scatterplot in matplotlip***

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.array([30,45,55,85,79,80,35,60,80,90])
woman=np.array([20,40,50,70,65,60,85,70,79,85])
man=np.array([25,35,45,60,55,60,80,70,75,80])

plt.scatter(x,woman,color='green',marker='^')
plt.scatter(x,man,color='red',marker='*')
plt.legend(['woman','man'],loc=6)
plt.show()
```

```python
# import matplotlib.pyplot as plt
import numpy as np
x=np.array([20,40,50,70,65,60,85,70,79,85])
y=np.array([15,35,45,60,55,60,80,70,75,80])
colors=np.array([10,40,50,70,65,60,85,70,79,85])    #### biger than array of color should be  useing cmap='any' display colors****
size=np.array([80,85,87,90,70,78,90,91,93,95])

plt.scatter(x,y,c=colors,cmap='gist_rainbow',s=size,alpha=0.5,label='man')

plt.legend(['man','woman'],loc=8)
plt.colorbar()
plt.show()
```

# stack plot in matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
x=[2,3,4,5,6]
y={"y1":[3,5,6,7,8],
    "y2":[7,8,5,2,7],
    "y3":[9,8,3,5,6],
    "y4":[4,5,7,8,9]}
c=["red","magenta","pink","hotpink"]
plt.stackplot(x,y.values(),colors=c, labels=y.keys())
plt.legend(loc=2)
plt.show() 
```

# fill_between plot in matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
a=np.linspace(0,2*3.14,70)
b=np.sin(a)

plt.fill_between(a,b,0,where=(a>1.5)&(a<=3.15),color="r",alpha=0.5)

plt.plot(a,b)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
a=np.linspace(0,2*3.14,70)
b=np.cos(a)

plt.fill_between(a,b,0,where=(a>1.5)&(a<=4.65),color="r",alpha=0.5)

plt.plot(a,b)

plt.show()
```
