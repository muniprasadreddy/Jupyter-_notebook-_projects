```python
import pandas as pd

def handle_null_values(data):
    """
    Treat null values in a DataFrame by filling numerical columns with median
    and categorical columns with mode.

    Parameters:
    df (DataFrame): Input DataFrame with null values.

    Returns:
    DataFrame: DataFrame with null values treated.
    """
    for column in data.columns:
        if data[column].dtype == 'object':  # Categorical columns
            mode_val = data[column].mode()[0]
            data[column].fillna(mode_val, inplace=True)
        elif data[column].dtype == 'category':  # Categorical columns
            mode_val = data[column].mode()[0]
            data[column].fillna(mode_val, inplace=True)
        else:  # Numerical columns
            median_val = data[column].median()
            data[column].fillna(median_val, inplace=True)

    return data

# Treat null values
null_vaules_data_processed = handle_null_values(data)

# Check if null values are handled
print(null_vaules_data_processed.isnull().sum())
```

```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd

def label_encode_columns(data):
    """
    Apply label encoding to all object dtype columns in a DataFrame.

    Parameters:
    df (DataFrame): Input DataFrame.

    Returns:
    DataFrame: DataFrame with label encoding applied to object dtype columns.
    """
    label_encoded_data = data.copy()  # Create a copy to avoid modifying the original DataFrame

    for column in label_encoded_data.columns:
        if (label_encoded_data[column].dtype == 'object') or (label_encoded_data[column].dtype == 'category') :
            label_encoder = LabelEncoder()
            label_encoded_data[column] = label_encoder.fit_transform(label_encoded_data[column])
    return label_encoded_data

# Apply label encoding to all object dtype columns
data_encoded = label_encode_columns(null_vaules_data_processed)

# Check the result
data_encoded.head()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
from sklearn.preprocessing import OneHotEncoder

# Assuming you have your data loaded into a DataFrame called 'data'
# Replace 'TARGET' with the name of your target column
X = data.drop(columns=['TARGET'])
y = data['TARGET']

# One-hot encode categorical columns
categorical_cols = X.select_dtypes(include=['object', 'category']).columns
X_encoded = pd.get_dummies(X, columns=categorical_cols, drop_first=True)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)

# Initialize the Decision Tree classifier
dt_classifier = DecisionTreeClassifier()

# Train the classifier on the training data
dt_classifier.fit(X_train, y_train)

# Predict labels on the test set
y_pred = dt_classifier.predict(X_test)

# Calculate confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(conf_matrix)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Generate classification report
class_report = classification_report(y_test, y_pred)
print("Classification Report:")
print(class_report)

# Perform cross-validation
cross_val_scores = cross_val_score(dt_classifier, X_encoded, y, cv=5)
print("Cross-Validation Scores:", cross_val_scores)
print("Mean Cross-Validation Score:", cross_val_scores.mean())
```
